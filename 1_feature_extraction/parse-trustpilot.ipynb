```python
import pandas as pd
import numpy as np
import json
```

```python
def load_json(path):
    f = open(path, "r")
    json_loaded = json.load(f)
    f.close()
    return json_loaded
```

```python
trustpilot_json = load_json("trustpilot.json")
trustpilot_json[0]
```

```python
def safe_execute(default, exception, line_to_exec):
    try:
        return line_to_exect
    except exception:
        return default
```

```python
trustpilot = []
for i in range(len(trustpilot_json)):
    
    domain = trustpilot_json[i]['domain']

    try:
        display_name = trustpilot_json[i]['general']['displayName']
    except:
        display_name = ''
        
    try:
        nb_reviews = trustpilot_json[i]['general']['numberOfReviews']
    except:
        try:
            nb_reviews = trustpilot_json[i]['agg_rating']['reviewCount']
        except:
            nb_reviews = ''
        
        
    try:
        trust_score = trustpilot_json[i]['general']['trustScore']
    except:
        try:
            trust_score = trustpilot_json[i]['agg_rating']['ratingValue']
        except:
            trust_score = ''
        
        
    try:
        is_closed = trustpilot_json[i]['general']['isClosed']
    except:
        is_closed = ''

        
    try:
        email = trustpilot_json[i]['email']
    except:
        email = ''
        
        
    try:
        telephone = trustpilot_json[i]['telephone']
    except:
        telephone = ''
        
        
    try:
        best_rating = trustpilot_json[i]['agg_rating']['bestRating']
    except:
        best_rating = ''
        
        
        
    try:
        worst_rating = trustpilot_json[i]['agg_rating']['worstRating']
    except:
        worst_rating = ''
        
        
        
    try:
        industries = trustpilot_json[i]['industry']['categories']
        industries = ','.join(industries)
    except:
        industries = ''
        
        
    try:
        nb_stars = trustpilot_json[i]['industry']['numberOfStars']
    except:
        nb_stars = ''
    
    trustpilot.append({
        'domain':domain,
        'display_name':display_name,
        'nb_reviews':nb_reviews,
        'trust_score':trust_score,
        'is_closed':is_closed,
        'email':email,
        'telephone':telephone,
        'best_rating':best_rating,
        'worst_rating':worst_rating,
        'industries':industries,
        'nb_stars':nb_stars
    })
```

```python
trustpilot_df = pd.DataFrame(trustpilot)
trustpilot_df.to_csv('trustpilot.csv')
trustpilot_df.to_excel('trustpilot.xlsx')
```
